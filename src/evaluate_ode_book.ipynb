## 加载模型 加载输入输出

```python
model, conf = get_model_from_run(run_paths[0])
n_dims = conf.model.n_dims
batch_size = 1
n_points = conf.training.curriculum.points.end
data_sampler = get_data_sampler(conf.training.data, n_dims)
task_sampler = get_task_sampler(
    conf.training.task,
    n_dims,
    batch_size,
    **conf.training.task_kwargs
)
task = task_sampler()
xs = data_sampler.continuous_sample_xs(b_size=batch_size, n_points=n_points,use_h=False)
ys = task.evaluate(xs)

outputs = []
```

```python
task_sampler_acc =  get_task_sampler(
        "ode_ivp_case2BSL",
        n_dims,
        batch_size
)
task_acc = task_sampler_acc()
ys_acc = task_acc.evaluate(xs,method = "DOP853")
ys = ys_acc
```

```python
print(f"xs shape: {xs.shape}")
print(f"ys shape: {plot_ys[1].shape}")
print(f"model input expected shape: [batch_size, n_points, n_dims]")
```

```python
# 转换模型输出结果
all_errs = []
for yss in plot_ys:
    yss =  yss.cuda()
    output = model(xs, yss)
    steps = get_steps(conf.training.data, xs)
    
    # 创建mask矩阵，steps范围外设为0
    mask = torch.zeros_like(output)
    for i in range(output.shape[0]):
        for j in range(output.shape[1]):
            mask[i, j, :steps[i, j]] = 1

    diff = torch.abs(output - yss) * mask
    error = torch.sum(diff, dim=-1) / steps
    
    all_errs.append(error)
    
```
