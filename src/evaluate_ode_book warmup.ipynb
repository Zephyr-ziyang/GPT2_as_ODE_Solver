```python
plot_ys = []
plot_ys.extend(outputs)
plot_labels = []
plot_labels.extend(output_labels)
```

```python
# 转换模型输出结果
all_errs = []
error_range_list = []  # 存储每个数据集的误差范围[min, max]

for k in range(len(plot_labels)):
    shorts_error = []
    current_pred = plot_ys[k]
    
    # 为每个step初始化存储min和max误差
    step_min_errors = [[] for _ in range(xs.shape[1])]  # 每个step的最小误差
    step_max_errors = [[] for _ in range(xs.shape[1])]  # 每个step的最大误差
    
    for i in range(xs.shape[0]):
        # 一个task
        errs = []
        
        for j in range(xs.shape[1]):
            # 一个step
            eval_y = current_pred[i, j]
            steps, t_eval, paras = get_paras(conf.training.data, xs, ys, eval_y, i, j)
            
            diff = abs(eval_y.cpu().numpy()-ys[i, j].cpu().numpy())
            err = np.mean(diff[:steps])
            errs.append(err)
            
            # 记录每个step的误差
            step_min_errors[j].append(np.min(diff[:steps]))
            step_max_errors[j].append(np.max(diff[:steps]))
            
        shorts_error.append(errs)
    
    # 计算每个step的最小和最大误差
    min_errors = [np.min(errs) for errs in step_min_errors]
    max_errors = [np.max(errs) for errs in step_max_errors]
    
    
    all_errs.append(shorts_error)
    error_range_list.append([min_errors, max_errors])  # 保存当前模型的误差范围
```

```python
global_errs_list = []
relative_errs_list = []
for short_errs in all_errs:
    global_errs = []
    relative_errs = []
    for i in range(len(short_errs[0])):
        vec_norm = np.linalg.norm([err[i] for err in short_errs])
        global_errs.append(vec_norm)
        true_norm = np.linalg.norm([ys[j,i].cpu().numpy() for j in range(ys.shape[0])])
        relative_errs.append(vec_norm / true_norm if true_norm != 0 else 0)
    global_errs_list.append(global_errs)
    relative_errs_list.append(relative_errs)
```
